## online payemnt fraud detection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## import online payment fraud dataset

```python
df= pd.read_csv('Online Payments Fraud Detection Dataset.csv')
df.head(10)
```

```python
# df['isFraud'].value_counts()
df["type"].value_counts()
```

## Normalizing data

```python
x = df[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]]
y = df["isFraud"]
```

```python
from sklearn.preprocessing import LabelEncoder
x = x.apply(LabelEncoder().fit_transform)
```

## Split train\test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
x_test_scaled = pd.DataFrame(scaler.transform(x_test), columns=x_test.columns)
```

## Prediction with Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train_scaled, y_train)
y_pred = dt.predict(x_test)
```

## Calculating Accuracy, Precision, Recall, Confusion Matrix

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("\nConfusion Matrix:\n", cm)
```

## User Inputs

```python
mappings = {
    "type": {
        "CASH_OUT": 1,
        "PAYMENT": 2,
        "CASH_IN": 3,
        "TRANSFER": 4,
        "DEBIT": 5
    }
}
```

```python
user_input = {
    "type": input("Transaction Type: "),
    "amount": float(input("Amount: ")),
    "oldbalanceOrg": float(input("Old Balance Origin: ")),
    "newbalanceOrig": float(input("New Balance Origin: "))
}
for col, mapping in mappings.items():
    if col in user_input:
        user_input[col] = mapping.get(user_input[col], 0) 
columns = x_train.columns  
user_df = pd.DataFrame([user_input], columns=columns)
user_scaled = pd.DataFrame(scaler.transform(user_df), columns=columns)
pred = dt.predict(user_scaled)[0]
label_map = {0: "No Fraud", 1: "Fraud"}
print("Prediction:", label_map[pred])
```
